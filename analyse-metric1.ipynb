# Metric 1

```python
from dangerosite import dangerosite
# would the player's team not affect the result of the function
df_pass = df[df['pass_available']==True]
df_pass['xT1'] = df_pass.apply(lambda row: dangerosite(row['location_x'],row['location_y']), axis = 1)
df_pass['xT2'] = df_pass.apply(lambda row: dangerosite(row['pass_end_location_x'],row['pass_end_location_y']), axis = 1)
df_pass['xT'] = df_pass['xT2'] - df_pass['xT1']
```

```python
df_pass_xt = df_pass[['player','xT']]
df_pass_xt = df_pass_xt[df_pass_xt['xT']>=0]
df_pass_xt.groupby('player').sum()
```
